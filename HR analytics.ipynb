## Data preprocessing

```python
# filling missing values
train_data['education'].fillna(train_data['education'].mode()[0], inplace=True)
# as most employees whose previous year ratings are missing are freshers with length of service 1.
train_data['previous_year_rating'].fillna(2, inplace=True)
```

```python
# filling missing values
test_data['education'].fillna(test_data['education'].mode()[0], inplace=True)
# as most employees whose previous year ratings are missing are freshers with length of service 1.
test_data['previous_year_rating'].fillna(2, inplace=True)
```

```python
test_data.isnull().any()
```

```python
train_data.isnull().any()
```

```python
train_data = train_data.drop(['employee_id'],axis=1)
train_data.columns
```

```python
emp_id = test_data['employee_id']
test_data= test_data.drop(['employee_id'], axis=1)
test_data.columns
```

```python
#defining test sets
x_test = test_data
x_test.columns
```

```python
x_test =pd.get_dummies(x_test)
x_test.columns
```

```python
# splitting the train set into dependent and independent sets
x= train_data.iloc[:,:-1]
y= train_data.iloc[:,-1]
```

```python
#one hot encoding
x =pd.get_dummies(x)
x.columns
```

```python
from imblearn.over_sampling import SMOTE

x_sample, y_sample = SMOTE().fit_resample(x, y.values.ravel())

x_sample = pd.DataFrame(x_sample)
y_sample = pd.DataFrame(y_sample)

# checking the sizes of the sample data
print("Size of x-sample :", x_sample.shape)
print("Size of y-sample :", y_sample.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_valid, y_train, y_valid= train_test_split(x_sample, y_sample,test_size=0.2, random_state=0)
print("Shape of x_train: ", x_train.shape)
print("Shape of x_valid: ", x_valid.shape)
print("Shape of y_train: ", y_train.shape)
print("Shape of y_valid: ", y_valid.shape)
```

```python
from sklearn.preprocessing import StandardScaler

sc= StandardScaler()
x_train =sc.fit_transform(x_train)
x_test= sc.fit_transform(x_test)
x_valid= sc.fit_transform(x_valid)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = None)
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
#x_valid = pca.transform(x_valid)
```

```python
import sys
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import average_precision_score

rfc = RandomForestClassifier()
rfc.fit(x_train, y_train.values.ravel())

rfc_pred = rfc.predict(x_test)

print("Training Accuracy :", rfc.score(x_train, y_train))
```

```python
print("Validation Accuracy :", rfc.score(x_valid, y_valid))
```

### XGBoost classifier

```python
from xgboost.sklearn import XGBClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import average_precision_score

xgb= XGBClassifier()
xgb.fit(x_train,y_train.values.ravel())

xgb_pred = xgb.predict(x_test)
print("Training Accuracy: ", xgb.score(x_train,y_train))
```

```python
print("Validation Accuracy: ",xgb.score(x_valid, y_valid))
```

### Light Gradient Boosting Classifier

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import average_precision_score

lgb= LGBMClassifier()
lgb.fit(x_train,y_train.values.ravel())

lgb_pred = lgb.predict(x_test)
print("Training Accuracy: ", lgb.score(x_train,y_train))
```

```python
print("Validation Accuracy: ", lgb.score(x_valid,y_valid))
```

## ExtraTrees Classifier

```python
from sklearn.ensemble import ExtraTreesClassifier

etc= ExtraTreesClassifier()
etc.fit(x_train,y_train.values.ravel())

etc_pred = etc.predict(x_test)
print("Training Accuracy: ", etc.score(x_train,y_train))
```

```python
print("Validation Accuracy: ", etc.score(x_valid,y_valid))
```

## CatBoost Classifier

```python
from catboost import CatBoostClassifier

cb= CatBoostClassifier(learning_rate = 0.04)
cb.fit(x_train,y_train.values.ravel())

cb_pred = cb.predict(x_test)
print("Training Score: ", cb.score(x_train,y_train))
```

```python
print("Validation Score: ", cb.score(x_valid,y_valid))
```
